# Generate ImageNet Images

## In this notebook I use the DCGAN trained using ImageNet to generate images

```python
import tensorflow as tf
import numpy as np
import tensorflow.keras.layers as layers
import matplotlib.pyplot as plt
import numpy as np
import PIL
from IPython import display
import tensorflow_datasets as tfds
import warnings
warnings.filterwarnings("ignore")
```

# Load Generator

```python
Generator = tf.keras.models.load_model('ImageNetFinal/gen_weights/gen_model_epoch_0044.h5')
```

```python
Generator.input
```

# Check Output

```python
rand = tf.random.normal([9, 128], stddev=0.2)
predictions = Generator(rand)
predictions = predictions.numpy()
fig = plt.figure(figsize=(8, 8))

for i in range(predictions.shape[0]):
    plt.subplot(3, 3, i + 1)
    plt.imshow((predictions[i, :, :, :] * 127.5 + 127.5).astype(np.uint8))
    plt.axis('off')
```

# Generate Images

```python
rand = tf.random.normal([100, 128], stddev=0.2)
predictions = Generator(rand)
predictions = predictions.numpy()
fig = plt.figure(figsize=(18, 18))

for i in range(predictions.shape[0]):
    plt.subplot(10, 10, i + 1)
    plt.imshow((predictions[i, :, :, :] * 127.5 + 127.5).astype(np.uint8))
    plt.axis('off')
fig.tight_layout()
fig.savefig('CIFAR-10GeneratedImages.png')
```
